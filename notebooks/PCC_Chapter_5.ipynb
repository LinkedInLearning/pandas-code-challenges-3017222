# Grouping and aggregation

```python
import pandas as pd
import numpy as np
```

##  amounts for each demographic based on marital status

```python
df = pd.read_csv("./data/marketing_campaign.csv", sep='\t')
df.info()
```

```python
amount_bought = [
        'MntWines',
        'MntFruits',
        'MntMeatProducts',
        'MntFishProducts',
        'MntSweetProducts',
        'MntGoldProds'
    ]
df.groupby(['Marital_Status']).mean()[amount_bought]
```

## 05_02 - Grouping by multiple columns

```python
df.groupby(['Education', 'Marital_Status'])['Income'].agg(['median', 'mean'])
```

## 05_03 - Applying a custom aggregate function

```python
df
```

```python
def top(df, n=5, column='NumWebPurchases'):
    return df.sort_values(by=column)[-n:]

    
```

```python

visits = ['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']
df.groupby(['Education']).apply(top)[visits]
```

## 05_04 - Filtering rows based on statistics created groups

```python
stocks = pd.read_csv("./stock_data.csv", parse_dates=True, index_col="Unnamed: 0")
stocks.head()
```

```python
stocks.info()
```

```python
##calculate returns
rets = stocks.pct_change().dropna()
rets
```

```python
get_year = lambda x: x.year
by_year = rets.groupby(get_year).sum()*100
by_year
```
