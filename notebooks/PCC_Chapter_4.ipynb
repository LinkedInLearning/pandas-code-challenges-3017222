## 04_02 Chained conditional statements
- Extract all the customers who don't have a partner and are born after 1990.

```python
df.Marital_Status.value_counts()
```

## 04_03 - Using bitwise operations

- Extract all customers who are active online.
- Extract all customers who are active offline and create a new column and store the results.


### Add new columns for active buyers

```python
df.loc[(df['NumWebPurchases'] > 10) & (df['NumDealsPurchases'] > 5)]
```

```python
##The product team wants to develop the roadmap for the online platform. 
##And for that they want to study people who often buy online on their platform.


df.loc[(df['NumWebPurchases'] > 5) | (df['NumDealsPurchases'] > 5), 'ActiveOnlineBuyers'] = 'Active'
df.loc[(df['NumWebPurchases'] < 5) & (df['NumDealsPurchases'] < 5), 'ActiveOnlineBuyers'] = 'Not Active'

```
