## Calculate yearly stock price returns

```python
import pandas as pd
```

```python
stocks = pd.read_csv("./data/stock_data.csv",
                index_col="Unnamed: 0",
                parse_dates=True)

stocks.head()
```

```python
stocks
```

```python
##calculate daily return

rets = stocks.pct_change().dropna()
rets
```

```python
get_year = lambda x: x.year
by_year_stocks = rets.groupby(get_year).sum()*100
by_year_stocks
```
