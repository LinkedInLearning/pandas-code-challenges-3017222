# Grouping data

```python
import pandas as pd
```

```python
df = pd.read_csv("./data/marketing_campaign.csv", sep='\t')
df.info()
```

## Average amount of each product bought by each group of customers

```python
amt_bought = [
    'MntWines',
    'MntFruits',
    'MntMeatProducts',
    'MntFishProducts',
    'MntSweetProducts',
    'MntGoldProds'
]
```

```python
df.groupby(['Marital_Status']).mean()[amt_bought]
```

## Grouping by multiple columns

```python
df.groupby(['Education', 'Marital_Status'])['Income'].agg(['median', 'mean'])
```

## Applying a custom aggregate function

```python
df
```

```python
def top(df, n=5, column="NumWebPurchases"):
    return df.sort_values(by=column)[-n:]
```

```python
purchases = ['NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases']
df.groupby(['Education']).apply(top)[purchases]
```
