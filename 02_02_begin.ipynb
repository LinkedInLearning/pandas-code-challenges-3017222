## Subsetting rows from the dataframe using .loc[ ]

```python
df2 = camp_df.set_index("ID")
```

```python
df3 = df2.loc[7446 : 2114]
```

```python
newcols = [ x for x in list(df3.columns) if  x.startswith("Mnt")]
```

```python
df5 = df3[ newcols]
```
