# Conditional filtering

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./data/marketing_campaign.csv", sep='\t')
df.head()
```

```python
df.info()
```

## Conditional filtering

```python
df[df['Marital_Status'] == "Married"]
```

```python
df.loc[df['Marital_Status'] == "Married"]
```

## Chained condition

```python
df['Marital_Status'].value_counts()
```

```python
df.loc[(~df['Marital_Status'].isin(['Married', 'Together'])) & (df['Year_Birth'] > 1990)]
```

## Using mutliple conditions to create new features

```python
df.loc[(df['NumWebPurchases'] > 5) | (df['NumDealsPurchases'] > 5), 'ActiveOnlineBuyers'] = "Active"
df.loc[(df['NumWebPurchases'] < 5) & (df['NumDealsPurchases'] < 5), 'ActiveOnlineBuyers'] = "Not Active"
```
